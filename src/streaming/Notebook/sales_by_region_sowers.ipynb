```python
import pandas as pd

df = pd.read_csv("../../../data/sales.csv")

df.head()
```

```python
region_counts = (
    df.groupby("region_id")
    .size()
    .reset_index(name="order_count")
    .sort_values("order_count", ascending=False)
)

region_counts
```

```python
import matplotlib.pyplot as plt

ax = region_counts.plot(
    x="region_id",
    y="order_count",
    kind="bar",
    color="turquoise",
    edgecolor="black",
    linewidth=1.2,
    legend=False,
    figsize=(8, 5),
)

plt.title("Order Count by Region", fontsize=16, fontweight="bold")
plt.xlabel("Region", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)

for container in ax.containers:
    ax.bar_label(container)

plt.ylim(0, 50)

plt.tight_layout()

plt.savefig(
    "../../../docs/images/orders_by_region_sowers.png", dpi=300, bbox_inches="tight"
)

plt.show()
```
